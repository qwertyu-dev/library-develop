```python
import pandas as pd

import numpy as np

# データフレームの作成
def create_dataframe(row_count):
    # データの生成
    df = pd.DataFrame({
        "申請元情報": ["人事"] * row_count,
        "種類": ["国企"] * row_count,
        "対象": ["関連"] * row_count,
        "部門コード": [f"{i+1:03}" for i in range(row_count)],
        "親部店コード": [f"{100 + i:03}" for i in range(row_count)],
        "部店コード": [f"{200 + i:03}" for i in range(row_count)],
        "部店名称": [f"部店{i+1}" for i in range(row_count)],
        "課/エリアコード": [f"{300 + i:03}" for i in range(row_count)],
        "課/エリア名称": [f"課エリア{i+1}" for i in range(row_count)],
        "課/エリア名称（英語）": [f"Area {i+1}" for i in range(row_count)],
        "常駐部店コード": [f"{400 + i:03}" for i in range(row_count)],
        "常駐部店名称": [f"常駐部店{i+1}" for i in range(row_count)],
        "共通認証受渡し予定日": [f"2024/03/{i+1:02}" for i in range(row_count)],
        "拠点内営業部情報": [f"営業部{i+1}" for i in range(row_count)],
        "エリア情報（課Grコード）": [f"{500 + i:03}" for i in range(row_count)],
        "備考": [f"特記事項{i+1}" for i in range(row_count)],
        "部店カナ": [f"ブテン{i+1}" for i in range(row_count)],
        "課名称（カナ）": [f"カメイ{i+1}" for i in range(row_count)],
        "課名称（略称）": [f"課略称{i+1}" for i in range(row_count)],
        "BPR対象/対象外フラグ": np.random.choice(["対象", "対象外"], row_count),
        "出向リカバリフラグ": np.random.choice([0, 1], row_count),
    })
    return df

# CSVファイルへの保存
def save_to_csv(df, filename):
    df.to_csv(filename, index=False, encoding="utf-8_sig")

# 実行
row_count = 30
df = create_dataframe(row_count)
```
