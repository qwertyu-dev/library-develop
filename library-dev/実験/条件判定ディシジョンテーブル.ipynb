```python
import pandas as pd

# ディシジョンテーブルを読み取る
decision_table = pd.read_excel('decision_table.xlsx')
print('-'*80)
print(decision_table)

# データサンプルを読み取る
data_sample = pd.read_excel('data_sample.xlsx')
print('-'*80)
print(data_sample)

def check_condition(value, condition):
    if pd.isna(condition) or condition == 'any':
        return True
    elif isinstance(condition, str):
        if ',' in condition:
            return any(str(value) == str(c.strip()) for c in condition.split(','))
        else:
            return str(value) == str(condition)
    else:
        return value == condition

def evaluate_conditions(row, decision_table):
    for _, decision_row in decision_table.iterrows():
        if all(check_condition(row[col], decision_row[col]) for col in row.index):
            return decision_row['判定結果']
    return 'No Match'

# データサンプルの各行に対して条件を評価し、判定結果を追加する
data_sample['判定結果'] = data_sample.apply(lambda row: evaluate_conditions(row, decision_table), axis=1)

# 結果を表示する
print(data_sample)
```

```python
import pandas as pd

# ディシジョンテーブルを読み取る
decision_table = pd.read_excel('decision_table.xlsx')
print('-'*80)
print(decision_table)

# データサンプルを読み取る
data_sample = pd.read_excel('data_sample.xlsx')
print('-'*80)
print(data_sample)

def check_condition(value, condition):
    # ディシジョンテーブルでany設定箇所は常にTrueで返す(評価しない)
    if pd.isna(condition) or condition == 'any':
        return True
    elif isinstance(condition, str):
        # 文字列で条件設定→ 1,2,3 のようなor条件設定
        # Excel→DataFrame取り込み時にはstr型となる
        if ',' in condition:
            return str(value) in [str(c.strip()) for c in condition.split(',')]
        else:
            return str(value) == str(condition)
    else:
        return value == condition

def evaluate_conditions(row, decision_table):
    for _, decision_row in decision_table.iterrows():
        if all(check_condition(row[col], decision_row[col]) for col in row.index):
            return decision_row['判定結果']
    return 'No Match'

# データサンプルの各行に対して条件を評価し、判定結果を追加する
data_sample['判定結果'] = data_sample.apply(lambda row: evaluate_conditions(row, decision_table), axis=1)

# 結果を表示する
print('-'*80)
print(data_sample)
```
